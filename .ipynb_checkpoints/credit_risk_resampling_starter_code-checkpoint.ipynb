```python
print(len(columns))
```

```python
# Load the data
# file_path = Path('LoanStats_2019Q1.csv.zip')
# df = pd.read_csv(file_path, skiprows=1)[:-2]
loans_df = df.loc[:, columns].copy()
loans_df.head()
```

```python
# Drop the null columns where all values are null
loans_df = loans_df.dropna(axis='columns', how='all')
loans_df.head()
```

```python
# Drop the null rows
loans_df = loans_df.dropna()
loans_df.info()
```

```python
# Remove the `Issued` loan status
issued_mask = loans_df['loan_status'] != 'Issued'
loans_df = loans_df.loc[issued_mask]
loans_df['loan_status'].value_counts()
```

```python
# Convert the target column values to low_risk and high_risk based on their values
x = {'Current': 'low_risk'}   
loans_df = loans_df.replace(x)
loans_df['loan_status'].value_counts()
```

```python
x = dict.fromkeys(['Late (31-120 days)', 'Late (16-30 days)', 'Default', 'In Grace Period'], 'high_risk')    
loans_df = loans_df.replace(x)
loans_df['loan_status'].value_counts()
```

```python
loans_df.head()
```

```python
# convert interest rate to numerical
loans_df['int_rate'] = loans_df['int_rate'].str.replace('%', '')
loans_df['int_rate'] = loans_df['int_rate'].astype('float') / 100
loans_df.head()
```

```python
loans_df.info()
```

```python
loans_df.reset_index(inplace=True, drop=True)

loans_df.head()
```

```python
# what are our categorical value columns
columns_obj = loans_df.select_dtypes(include='object').columns
print(len(columns_obj))
```

```python
# get dummies method
columns_obj
```

```python
loans_df['home_ownership'].value_counts()
```

```python
for column in loans_df[columns_obj]:
    print(column, "\n", loans_df[column].value_counts())
    print("-----------------------------------------------")
```

```python
# drop colums that have one value in value_counts()
loans_df = loans_df.drop(columns=['debt_settlement_flag', 'hardship_flag', 'pymnt_plan'])
loans_df.shape
```

```python
loans_df.head()
```

```python
# save our loans_df as clean_loans_df csv file
clean_loans_df = loans_df
clean_loans_df.to_csv("./Resources/clean_loans.csv",index=False)
```

```python
# read in the csv file
df = pd.read_csv("./Resources/clean_loans.csv")
df.head()
```

# Split the Data into Training and Testing

```python
# Create our features
X = df.drop(columns='loan_status')
X.info()
```

```python
# Create our target
y = df['loan_status']
y.value_counts()
```

```python
# this is our imbalanced problem
print((347/68470)*100)
```

```python
X.describe()
```

```python
# Check the balance of our target values
# y['loan_status'].value_counts()
```

```python
# use get dummies to transform object types to integers
X_encoded = pd.get_dummies(X)
X_encoded.shape
```

```python
# all our columns in our featured matrix (X) are numeric
X_encoded.describe()
```

```python
X.shape
```

```python
y.shape
```

### Use module 17.6.4 to scale data in X_encoded

```python
# scaleX_encoded 
data_scaler = StandardScaler()
```

```python
X_encoded_scaled = data_scaler.fit_transform(X_encoded)
X_encoded_scaled[:1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded_scaled, y, random_state = 1)
```

```python
y_train.value_counts()
```

```python
print((246/51366)*100)
```

```python
# get a baseline of performance using logistic regression
```

```python
# Instantiate a Logistic Regression Model
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
# git model with our data (training)
classifier.fit(X_train, y_train)
```

```python
# create predictions using predict() methods
y_pred = classifier.predict(X_test)
y_pred
```

```python
pd.DataFrame({"Prediction": y_pred, "Actual": y_test})
```

```python
# calculate model accuracy
accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = classifier.classes_)
disp.plot()
```

```python
# view the classification report
report = classification_report(y_test, y_pred)
print(report)
```
